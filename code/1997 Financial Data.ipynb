```python
import pandas as pd

def load_1997_individual_unit(fpath: str) -> pd.DataFrame:
    """
    Load a single 1997 CoG Individual Unit item codes file.
    Fixed 35-character ASCII layout:
      [0:14]  Government Code
      [14:17] Item Code
      [17:29] Amount in $1000
      [29:31] Survey Year
      [31:33] Year of Data
      [33:35] Origin Code
    """
    records = []
    with open(fpath, "r", encoding="ascii", errors="replace") as f:
        for line in f:
            line = line.rstrip("\n")
            if len(line) < 35:
                continue
            gov_code     = line[0:14]
            item_code    = line[14:17].strip()
            amount_raw   = line[17:29].strip()
            survey_year  = line[29:31].strip()
            year_of_data = line[31:33].strip()
            origin_code  = line[33:35].strip()

            try:
                amount = int(amount_raw)
            except ValueError:
                amount = None

            records.append({
                "GOVT_CODE":    gov_code,
                "STATE_CODE":   gov_code[0:2],
                "TYPE_CODE":    gov_code[2:3],
                "COUNTY_CODE":  gov_code[3:6],
                "UNIT_CODE":    gov_code[6:9],
                "SUP_CODE":     gov_code[9:12],
                "SUB_CODE":     gov_code[12:14],
                "ITEM_CODE":    item_code,
                "AMOUNT":       amount,
                "SURVEY_YEAR":  survey_year,
                "YEAR_OF_DATA": year_of_data,
                "ORIGIN_CODE":  origin_code,
            })

    df = pd.DataFrame(records)
    print(f"Loaded {len(df):,} records")
    print(f"Unique governments: {df['GOVT_CODE'].nunique():,}")
    print(f"Unique item codes:  {df['ITEM_CODE'].nunique():,}")
    print(f"State codes present: {sorted(df['STATE_CODE'].unique())}")
    return df

fpath = "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1997_Individual_Unit_file/97ItemCodes0500.txt"
df_1997 = load_1997_individual_unit(fpath)
df_1997.head(10)
```

```python
def load_1997_directory(fpath: str) -> pd.DataFrame:
    """
    Load 1997 CoG Individual Unit Directory file.
    Fixed-length ASCII, 165 characters per record.

    [0:2]    State code
    [2:3]    Type code
    [3:6]    County code
    [6:9]    Unit code
    [9:12]   SUP code
    [12:14]  Sub code
    [14:78]  Government Name
    [78:79]  Region Code
    [79:114] County Name
    [114:116] FIPS State code
    [116:119] FIPS County code
    [119:124] FIPS Place code
    [124:128] FIPS MSA code
    [128:130] FIPS CMSA code
    [130:139] Population
    [139:141] Population Year
    [141:148] Enrollment
    [148:150] School Level code
    [150:157] Weight
    [157:161] Fiscal Year
    [161:163] Survey Year
    [163:165] Status Flag
    """
    records = []
    with open(fpath, "r", encoding="ascii", errors="replace") as f:
        for line in f:
            line = line.rstrip("\n")
            if len(line) < 14:
                continue
            records.append({
                "GOVT_CODE":        line[0:14],
                "STATE_CODE":       line[0:2],
                "TYPE_CODE":        line[2:3],
                "COUNTY_CODE":      line[3:6],
                "UNIT_CODE":        line[6:9],
                "SUP_CODE":         line[9:12],
                "SUB_CODE":         line[12:14],
                "GOVT_NAME":        line[14:78].strip(),
                "REGION_CODE":      line[78:79].strip(),
                "COUNTY_NAME":      line[79:114].strip(),
                "FIPS_STATE":       line[114:116].strip(),
                "FIPS_COUNTY":      line[116:119].strip(),
                "FIPS_PLACE":       line[119:124].strip(),
                "FIPS_MSA":         line[124:128].strip(),
                "FIPS_CMSA":        line[128:130].strip(),
                "POPULATION":       line[130:139].strip(),
                "POP_YEAR":         line[139:141].strip(),
                "ENROLLMENT":       line[141:148].strip(),
                "SCHOOL_LEVEL":     line[148:150].strip(),
                "WEIGHT":           line[150:157].strip(),
                "FISCAL_YEAR":      line[157:161].strip(),
                "SURVEY_YEAR":      line[161:163].strip(),
                "STATUS_FLAG":      line[163:165].strip(),
            })

    df = pd.DataFrame(records)
    print(f"Loaded {len(df):,} records")
    print(f"Unique governments: {df['GOVT_CODE'].nunique():,}")
    return df

fpath_dir = "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1997_Individual_Unit_file/97CensusIndUnitDirectory.txt"
df_1997_dir = load_1997_directory(fpath_dir)
df_1997_dir.head(10)
```

```python
df_1997_merged = df_1997.merge(
    df_1997_dir[["GOVT_CODE", "GOVT_NAME", "COUNTY_NAME", "FIPS_STATE", "FIPS_COUNTY", "FIPS_PLACE", "POPULATION", "POP_YEAR", "FISCAL_YEAR", "SURVEY_YEAR", "STATUS_FLAG"]],
    on="GOVT_CODE",
    how="left"
)

print(f"Rows: {len(df_1997_merged):,}")
print(f"Matched (GOVT_NAME not null): {df_1997_merged['GOVT_NAME'].notna().sum():,}")
print(f"Unmatched: {df_1997_merged['GOVT_NAME'].isna().sum():,}")
df_1997_merged.head(10)
```

```python
df_1997_places = df_1997_merged[df_1997_merged["TYPE_CODE"].isin(["2", "3"])].copy()

target_codes = [
    "B01", "B21", "B22", "B30", "B42", "B46", "B50", "B59", "B79", "B80", "B89",
    "T01", "T09", "T10", "T11", "T13", "T16", "T19", "T20", "T21", "T22", "T27", "T28", "T29"
]

# Filter to target codes only
df_1997_filtered = df_1997_places[df_1997_places["ITEM_CODE"].isin(target_codes)].copy()

# Sum B codes and T codes per government
df_1997_agg = df_1997_filtered.groupby("GOVT_CODE").apply(
    lambda x: pd.Series({
        "total_fed_transfers": x.loc[x["ITEM_CODE"].str.startswith("B"), "AMOUNT"].sum(),
        "total_tax_rev":       x.loc[x["ITEM_CODE"].str.startswith("T"), "AMOUNT"].sum(),
    })
).reset_index()

# Bring back directory fields
dir_fields = df_1997_dir[["GOVT_CODE", "GOVT_NAME", "COUNTY_NAME", "FIPS_STATE", "FIPS_COUNTY", "FIPS_PLACE", "POPULATION", "FISCAL_YEAR"]].drop_duplicates("GOVT_CODE")
df_1997_agg = df_1997_agg.merge(dir_fields, on="GOVT_CODE", how="left")

```

```python
# BEA deflation, Per capita, and multiplied by 1000 to get standardized values

deflator_1997 = 100.000 / 58.439

df_1997_agg["POPULATION"] = pd.to_numeric(df_1997_agg["POPULATION"], errors="coerce")

df_1997_agg["total_fed_transfers_pc"] = (df_1997_agg["total_fed_transfers"] * 1000 * deflator_1997) / df_1997_agg["POPULATION"]
df_1997_agg["total_tax_rev_pc"]       = (df_1997_agg["total_tax_rev"]       * 1000 * deflator_1997) / df_1997_agg["POPULATION"]

print(df_1997_agg[["GOVT_CODE", "GOVT_NAME", "POPULATION", "total_fed_transfers", "total_fed_transfers_pc", "total_tax_rev", "total_tax_rev_pc"]].head(10).to_string())
```
