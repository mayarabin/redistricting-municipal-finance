```python
import pandas as pd

def load_cog_1992(filepath):
    """
    Load 1992 Census of Governments Individual Unit File.
    Fixed-width ASCII, 1-indexed positions converted to 0-indexed slices.
    """
    colspecs = [
        (0, 14),   # Government Code (positions 01-14)
        (14, 17),  # Item Code (positions 15-17)
        (17, 29),  # Amount in $1000 (positions 18-29)
        (29, 31),  # Survey Year (positions 30-31)
        (31, 33),  # Year of Data (positions 32-33)
        (33, 35),  # Origin Code (positions 34-35)
    ]

    col_names = [
        'govt_code',
        'item_code',
        'amount',
        'survey_year',
        'year_of_data',
        'origin_code',
    ]

    df = pd.read_fwf(
        filepath,
        colspecs=colspecs,
        names=col_names,
        dtype={
            'govt_code':    str,
            'item_code':    str,
            'amount':       'Int64',
            'survey_year':  str,
            'year_of_data': str,
            'origin_code':  str,
        },
        header=None,
    )

    # Parse Government Code sub-fields
    df['state_code']  = df['govt_code'].str[0:2]
    df['type_code']   = df['govt_code'].str[2:3]
    df['county_code'] = df['govt_code'].str[3:6]
    df['unit_id']     = df['govt_code'].str[6:9]   # 3-digit place unit code
    df['sup_code']    = df['govt_code'].str[9:12]
    df['sub_code']    = df['govt_code'].str[12:14]

    return df


df92 = load_cog_1992(
    "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1992_Individual_Unit_File/92CensusIndUnit.txt"
)

print(df92.shape)
print(df92.head(10))
print(df92['item_code'].value_counts().head(20))
```

```python

def load_cog_1992_directory(filepath):
    colspecs = [
        (0,   2),    # State code        01-02
        (2,   3),    # Type code         03-03
        (3,   6),    # County code       04-06
        (6,   9),    # Unit code         07-09
        (9,   12),   # SUP code          10-12
        (12,  14),   # Sub code          13-14
        (14,  78),   # Government Name   15-78
        (78,  79),   # Region Code       79-79
        (79,  109),  # County Name       80-109
        (109, 111),  # FIPS State code   110-111
        (111, 114),  # FIPS County code  112-114
        (114, 119),  # FIPS Place code   115-119
        (119, 123),  # FIPS MSA code     120-123
        (123, 125),  # FIPS CMSA code    124-125
        (125, 134),  # Pop/Enroll/Func   126-134
        (134, 136),  # Pop/Enroll Year   135-136
        (136, 138),  # School Level      137-138
        (138, 145),  # NCES code         139-145
        (145, 151),  # Weight            146-150  ← codebook ends at 150, gap at 151-152
        (152, 156),  # Fiscal Year End   153-156
        (156, 186),  # Multi-County      157-186
        (186, 188),  # Reciprocal Weight 187-188
        (188, 189),  # Weekly Hours      189-189
        (189, 190),  # FT Pay Code       190-190
        (190, 191),  # PT Pay Code       191-191
        (191, 193),  # Months Teachers   192-193
        (193, 195),  # Months Admin      194-195
        (195, 197),  # Months Other      196-197
        (197, 199),  # Year of Data      198-199
        (199, 201),  # Year Dep School   200-201
        (201, 203),  # Year Retirement   202-203
        (203, 204),  # Finance Edit Flag 204-204
        (204, 206),  # Survey Form Type  205-206
    ]
    col_names = [
        'state_code',
        'type_code',
        'county_code',
        'unit_code',
        'sup_code',
        'sub_code',
        'govt_name',
        'region_code',
        'county_name',
        'fips_state',
        'fips_county',
        'fips_place',
        'fips_msa',
        'fips_cmsa',
        'pop_enroll_func_code',
        'pop_enroll_year',
        'school_level_code',
        'nces_code',
        'weight',
        'fiscal_year_end',
        'multi_county_code',
        'reciprocal_weight',
        'weekly_hours_code',
        'ft_pay_code',
        'pt_pay_code',
        'months_teachers_paid',
        'months_school_admin_paid',
        'months_school_other_paid',
        'year_of_data',
        'year_dependent_school',
        'year_retirement_data',
        'finance_edit_flag',
        'survey_form_type',
    ]

    df = pd.read_fwf(
        filepath,
        colspecs=colspecs,
        names=col_names,
        dtype=str,
        header=None,
    )

    return df


dir92 = load_cog_1992_directory(
    "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1992_Individual_Unit_File/92CensusIndUnitDirectory.txt"
)

print(dir92.shape)
print(dir92.head(5).to_string())
```

```python
dir92_slim = dir92[[
    'state_code',
    'type_code',
    'county_code',
    'unit_code',
    'sup_code',
    'sub_code',
    'govt_name',
    'region_code',
    'county_name',
    'fips_state',
    'fips_county',
    'fips_place',
    'fips_msa',
    'fips_cmsa',
    'pop_enroll_func_code',
]].copy()

print(dir92_slim.shape)
print(dir92_slim.head(10).to_string())
```

```python
# filters down to the municipality government types only
dir92_filt = dir92_slim[dir92_slim['type_code'].isin(['2', '3'])].copy()
df92_filt = df92[df92['type_code'].isin(['2', '3'])].copy()

# re creates the government code a unique code present in both the dir and df documents
dir92_filt['govt_code'] = (
    dir92_filt['state_code'] +
    dir92_filt['type_code'] +
    dir92_filt['county_code'] +
    dir92_filt['unit_code']
)

# merges the directory and financial documents together
df92_merged = df92_filt.merge(
    dir92_filt[['govt_code', 'govt_name', 'fips_state', 'fips_county', 'county_name', 'pop_enroll_func_code']],
    on='govt_code',
    how='left'
)
```

```python
# filters to the specific b, federal transfer, and t, local tax revenues, item codes from the overall dataset
b_codes = ['B01', 'B21', 'B22', 'B30', 'B42', 'B46', 'B50', 'B59', 'B79', 'B80', 'B89']
t_codes = ['T01', 'T09', 'T10', 'T11', 'T13', 'T15', 'T16', 'T19', 'T20', 'T21', 'T22', 'T27', 'T28', 'T29', 'T40', 'T99']

df92_b = df92_merged[df92_merged['item_code'].isin(b_codes)].copy()
df92_b['series'] = 'B'

df92_t = df92_merged[df92_merged['item_code'].isin(t_codes)].copy()
df92_t['series'] = 'T'

# adds the t and b codes together by municipality so that there are two financial lines per a city
df92_agg = pd.concat([df92_b, df92_t]).groupby(
    ['govt_code', 'govt_name', 'fips_state', 'fips_county', 'county_name', 'series', 'pop_enroll_func_code']
)['amount'].sum().reset_index()

# applied the bea deflator, multiplies by 1000, divides by population to get real per a capita financial numbers
df92_agg['population'] = df92_agg['pop_enroll_func_code'].str.strip().astype('Int64')


deflator_1992 = 100.0 / 51.978
df92_agg['amount_nominal'] = df92_agg['amount'] * 1000
df92_agg['amount_real'] = df92_agg['amount_nominal'] * deflator_1992
df92_agg['amount_real_pc'] = df92_agg['amount_real'] / df92_agg['population']
```

```python
PLACES_PATH = "/Users/mayarabin/Desktop/Thesis Files/places_panel_final.parquet"

# Load places panel
places = pd.read_parquet(PLACES_PATH)
```

```python
import re

SUFFIXES = {'city', 'town', 'village', 'township', 'borough', 'municipality',
            'plantation', 'grant', 'location', 'purchase'}

def normalize_with_suffix(name):
    if not isinstance(name, str):
        return ''
    return re.sub(r'[^a-z0-9 ]', '', name.lower().strip())

def normalize_stripped(name):
    if not isinstance(name, str):
        return ''
    tokens = re.sub(r'[^a-z0-9 ]', '', name.lower().strip()).split()
    tokens = [t for t in tokens if t not in SUFFIXES]
    return ' '.join(tokens)

# ── fips_state -> state_name mapping ──────────────────────────────────────
fips_to_state = (places[['PLACE_ID', 'state_name']]
                 .assign(fips=places['PLACE_ID'].astype(str).str[:2])
                 .drop_duplicates(subset='fips')
                 .set_index('fips')['state_name']
                 .to_dict())

# ── Filter places panel to 1990 decade ────────────────────────────────────
places_1990 = places[places['year'] == 1990].copy()
places_1990['PLACE_ID'] = places_1990['PLACE_ID'].astype(str).str.strip()
places_1990['is_mcd'] = places_1990['PLACE_ID'].str.len() == 10

# ── Build name lookup tables ───────────────────────────────────────────────
# MCD: keyed on (state, county_fips_3, normalized_name)
# non-MCD: keyed on (state, normalized_name)
lookup_mcd_with     = {}
lookup_mcd_stripped = {}
lookup_place_with     = {}
lookup_place_stripped = {}

for _, row in places_1990.iterrows():
    place_id = str(row['PLACE_ID']).strip()
    state    = row['state_name']
    if row['is_mcd']:
        county = place_id[2:5]
        lookup_mcd_with.setdefault((state, county, normalize_with_suffix(row['NAME'])), []).append(place_id)
        lookup_mcd_stripped.setdefault((state, county, normalize_stripped(row['NAME'])), []).append(place_id)
    else:
        lookup_place_with.setdefault((state, normalize_with_suffix(row['NAME'])), []).append(place_id)
        lookup_place_stripped.setdefault((state, normalize_stripped(row['NAME'])), []).append(place_id)

# ── Name match function ────────────────────────────────────────────────────
def match_name(row):
    fips_state  = str(row['fips_state']).strip().zfill(2)
    fips_county = str(row['fips_county']).strip().zfill(3)
    state       = fips_to_state.get(fips_state, '')
    raw_name    = row['govt_name']

    if not state or not isinstance(raw_name, str):
        return None

    # Try MCD lookup first (county-scoped)
    candidates = lookup_mcd_with.get((state, fips_county, normalize_with_suffix(raw_name)), [])
    if len(candidates) == 1:
        return candidates[0]

    candidates = lookup_mcd_stripped.get((state, fips_county, normalize_stripped(raw_name)), [])
    if len(candidates) == 1:
        return candidates[0]

    # Fall back to non-MCD lookup (state-scoped)
    candidates = lookup_place_with.get((state, normalize_with_suffix(raw_name)), [])
    if len(candidates) == 1:
        return candidates[0]

    candidates = lookup_place_stripped.get((state, normalize_stripped(raw_name)), [])
    if len(candidates) == 1:
        return candidates[0]

    return None

# ── Apply match ────────────────────────────────────────────────────────────
df92_agg['PLACE_ID'] = df92_agg.apply(match_name, axis=1)

matched   = df92_agg['PLACE_ID'].notna().sum()
unmatched = df92_agg['PLACE_ID'].isna().sum()
print(f"Matched: {matched:,}, Unmatched: {unmatched:,}, Rate: {matched/(matched+unmatched):.1%}")

# ── Merge with places panel ────────────────────────────────────────────────
merged_92 = df92_agg.merge(places_1990, on='PLACE_ID', how='left')
print(merged_92.shape)
print(merged_92.head(10).to_string())
```
