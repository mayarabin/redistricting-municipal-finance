```python
import pandas as pd

def load_cog_1992(filepath):
    """
    Load 1992 Census of Governments Individual Unit File.
    Fixed-width ASCII, 1-indexed positions converted to 0-indexed slices.
    """
    colspecs = [
        (0, 14),   # Government Code (positions 01-14)
        (14, 17),  # Item Code (positions 15-17)
        (17, 29),  # Amount in $1000 (positions 18-29)
        (29, 31),  # Survey Year (positions 30-31)
        (31, 33),  # Year of Data (positions 32-33)
        (33, 35),  # Origin Code (positions 34-35)
    ]

    col_names = [
        'govt_code',
        'item_code',
        'amount',
        'survey_year',
        'year_of_data',
        'origin_code',
    ]

    df = pd.read_fwf(
        filepath,
        colspecs=colspecs,
        names=col_names,
        dtype={
            'govt_code':    str,
            'item_code':    str,
            'amount':       'Int64',
            'survey_year':  str,
            'year_of_data': str,
            'origin_code':  str,
        },
        header=None,
    )

    # Parse Government Code sub-fields
    df['state_code']  = df['govt_code'].str[0:2]
    df['type_code']   = df['govt_code'].str[2:3]
    df['county_code'] = df['govt_code'].str[3:6]
    df['unit_id']     = df['govt_code'].str[6:9]   # 3-digit place unit code
    df['sup_code']    = df['govt_code'].str[9:12]
    df['sub_code']    = df['govt_code'].str[12:14]

    return df


df92 = load_cog_1992(
    "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1992_Individual_Unit_File/92CensusIndUnit.txt"
)

print(df92.shape)
print(df92.head(10))
print(df92['item_code'].value_counts().head(20))
```

```python

def load_cog_1992_directory(filepath):
    colspecs = [
        (0,   2),    # State code        01-02
        (2,   3),    # Type code         03-03
        (3,   6),    # County code       04-06
        (6,   9),    # Unit code         07-09
        (9,   12),   # SUP code          10-12
        (12,  14),   # Sub code          13-14
        (14,  78),   # Government Name   15-78
        (78,  79),   # Region Code       79-79
        (79,  109),  # County Name       80-109
        (109, 111),  # FIPS State code   110-111
        (111, 114),  # FIPS County code  112-114
        (114, 119),  # FIPS Place code   115-119
        (119, 123),  # FIPS MSA code     120-123
        (123, 125),  # FIPS CMSA code    124-125
        (125, 134),  # Pop/Enroll/Func   126-134
        (134, 136),  # Pop/Enroll Year   135-136
        (136, 138),  # School Level      137-138
        (138, 145),  # NCES code         139-145
        (145, 151),  # Weight            146-150  ← codebook ends at 150, gap at 151-152
        (152, 156),  # Fiscal Year End   153-156
        (156, 186),  # Multi-County      157-186
        (186, 188),  # Reciprocal Weight 187-188
        (188, 189),  # Weekly Hours      189-189
        (189, 190),  # FT Pay Code       190-190
        (190, 191),  # PT Pay Code       191-191
        (191, 193),  # Months Teachers   192-193
        (193, 195),  # Months Admin      194-195
        (195, 197),  # Months Other      196-197
        (197, 199),  # Year of Data      198-199
        (199, 201),  # Year Dep School   200-201
        (201, 203),  # Year Retirement   202-203
        (203, 204),  # Finance Edit Flag 204-204
        (204, 206),  # Survey Form Type  205-206
    ]
    col_names = [
        'state_code',
        'type_code',
        'county_code',
        'unit_code',
        'sup_code',
        'sub_code',
        'govt_name',
        'region_code',
        'county_name',
        'fips_state',
        'fips_county',
        'fips_place',
        'fips_msa',
        'fips_cmsa',
        'pop_enroll_func_code',
        'pop_enroll_year',
        'school_level_code',
        'nces_code',
        'weight',
        'fiscal_year_end',
        'multi_county_code',
        'reciprocal_weight',
        'weekly_hours_code',
        'ft_pay_code',
        'pt_pay_code',
        'months_teachers_paid',
        'months_school_admin_paid',
        'months_school_other_paid',
        'year_of_data',
        'year_dependent_school',
        'year_retirement_data',
        'finance_edit_flag',
        'survey_form_type',
    ]

    df = pd.read_fwf(
        filepath,
        colspecs=colspecs,
        names=col_names,
        dtype=str,
        header=None,
    )

    return df


dir92 = load_cog_1992_directory(
    "/Users/mayarabin/Desktop/Thesis Downloads V2/Small Town:City Economic Data/1992_Individual_Unit_File/92CensusIndUnitDirectory.txt"
)

print(dir92.shape)
print(dir92.head(5).to_string())
```

```python
dir92_slim = dir92[[
    'state_code',
    'type_code',
    'county_code',
    'unit_code',
    'sup_code',
    'sub_code',
    'govt_name',
    'region_code',
    'county_name',
    'fips_state',
    'fips_county',
    'fips_place',
    'fips_msa',
    'fips_cmsa',
]].copy()

print(dir92_slim.shape)
print(dir92_slim.head(10).to_string())
```

```python
dir92_filt = dir92_slim[dir92_slim['type_code'].isin(['2', '3'])].copy()
df92_filt = df92[df92['type_code'].isin(['2', '3'])].copy()

dir92_filt['govt_code'] = (
    dir92_filt['state_code'] +
    dir92_filt['type_code'] +
    dir92_filt['county_code'] +
    dir92_filt['unit_code']
)


df92_merged = df92_filt.merge(
    dir92_filt[['govt_code', 'govt_name', 'fips_state', 'fips_county', 'county_name']],
    on='govt_code',
    how='left'
)

df92_merged.head(50)
```
